## Model building

```python
import warnings
warnings.filterwarnings('ignore')
import statsmodels.api as sm
from sklearn.cluster import KMeans
```

```python
l2 = fdata.iloc[:,18:20:1]
```

```python
kmeans = KMeans(5)
kmeans.fit(l2)

identified_clusters =kmeans.fit_predict(l2)
identified_clusters = list(identified_clusters)
```

```python
fdata['loc_clusters'] = identified_clusters
```

```python
# visualizing the clusters
fig, axes = plt.subplots(1,1, figsize=(30,20))
plt.scatter(fdata['latitude'], fdata['longitude'], c=fdata['loc_clusters'])
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.title('KMeans Clustering')
plt.show
```

```python

# Ask user for city
input_city = input("Enter a city name: ")

# Find its cluster
cluster = fdata.loc[fdata['city'] == input_city, 'loc_clusters'].iloc[0]

# Get all cities in the same cluster
unique_cities = fdata.loc[fdata['loc_clusters'] == cluster, 'city'].unique()

print("\nCities in the same cluster as", input_city, ":")
for city in unique_cities:
    if city != input_city:
        print("-", city)

# Ask user once for city selection
selected_city = input("\nEnter the city name to view hotel details: ")

# Validate selection
if selected_city in unique_cities:
    selected_hotels = fdata.loc[fdata['city'] == selected_city,
                                ['hotel_facilities', 'hotel_category', 'hotel_star_rating']]
    print("\nHotel Details for", selected_city)
    print(pd.DataFrame(selected_hotels))
else:
    print("❌ Invalid city selection. Please choose from the available cities.")

```

## Evaluation metrics

```python
#Evaluation metrics
wcss = kmeans.inertia_
print("Within-Cluster Sum of Squares (WCSS):", wcss)
```

```python
from sklearn.metrics import silhouette_samples, silhouette_score
```

```python

labels = kmeans.labels_
data = l2

# Calculate the silhouette samples
silhouette_samples = silhouette_samples(data, labels)

# Calculate the overall silhouette score
silhouette_avg = silhouette_score(data, labels)
```
